```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(1000)
y = np.random.standard_normal(20)
```

```python
x = range(len(y))
plt.plot (x,y)
```

the magority of `ndarray` methods return an `ndarray`. they can be passed to plt with a method

```python
print(y.cumsum())
plt.plot(y.cumsum())
```

```python
#customized
plt.plot(y.cumsum())
plt.grid(True)
plt.axis('tight')
plt.axis('scaled')
```

```python
#customized
plt.figure (figsize= (7,4)) # figsize parameter of figure defines the size of the fig (w,h)

plt.plot(y.cumsum(), 'b', lw = 1.5)
plt.plot (y.cumsum(), 'ro')
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
```

two or more graphs

```python
np.random.seed(2000)
y = np.random.standard_normal((20, 2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(7,5)) #the parameter figsize() for plt.figure
plt.subplot(2,1,1)     #declaration (rows,columns,which)
plt.plot( y[:,0], 'b', lw=1.5, label = '1st')  #lw for linewidth parameter
plt.plot (y[:,0], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
plt.title('A Simple Plot')
plt.subplot(212)
plt.plot (y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot (y[:, 1], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
```
