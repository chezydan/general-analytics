```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head(3)
```

```python
#compute total consumption
drinks['Total'] = drinks[ ['beer_servings', 'spirit_servings','wine_servings'  ]].sum(axis=1)
drinks
```

```python
mskNozero=drinks['Total']!=0
mskNozero[:4]
```

```python
drinksActive=drinks[mskNozero]
drinksActive.head()
```

```python
beer= drinksActive.beer_servings
wine= drinksActive.wine_servings
np.corrcoef(drinksActive['beer_servings'], drinksActive.wine_servings)
```

```python
plt.figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(beer,wine)
plt.xlabel("beer consumption")
plt.ylabel("wine consumption")
```

```python
continentsSet=set(drinksActive.continent)
print(len(continentsSet))
continents=list(continentsSet)
```

```python
#get coors for a country
def getCoors(country='Argentina'):
    beer=drinksActive.beer_servings[drinksActive['country']==country]
    wine=drinksActive.wine_servings[drinksActive['country']==country]
    
    return (beer.iloc[0],wine.iloc[0])
print(getCoors("Chile"))
```

```python
(continents)
```

```python
colors=['black', 'blue', 'red', 'yellow', 'orange', 'green']
plt.figure(figsize=(10,8))
for continent, color in zip(continents,colors):
    mskContinent = drinksActive['continent']==continent
    plt.scatter(beer[mskContinent],wine[mskContinent], color=color)
    plt.annotate("IL", xy=getCoors("Israel"))
    

plt.xlabel("beer consumption")
plt.ylabel("wine consumption")
plt.show()
```

```python
drinksActive[drinksActive['continent']=='South America'].head()
```

```python
cont='Asia'
mskContinent = drinksActive['continent']==cont
country='Uruguay'
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(beer[mskContinent],wine[mskContinent], color='lightblue')

plt.annotate("UR", xy=getCoors(country))
#plt.annotate("CL", xy=getCoors("Chile"))

def locatecountry(name, code="x"):
    print(code)
    plt.annotate(code, xy=getCoors(name))
    print(getCoors(name))

locatecountry("Israel", "IL")  
locatecountry("Brazil", "BR")  

plt.show()
```
